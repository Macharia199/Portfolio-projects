```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']
pd.options.mode.chained_assignment=None

#read data
df=pd.read_csv(r'C:\Users\NEWTON\Desktop\career2\movies.csv')
```

```python
df
```

```python
#check for missing data
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{}-{}%'.format(col,round(pct_missing*100)))
```

```python
print(df.dtypes)
```

```python
#checking for outliers
df.boxplot(column=['gross'])
```

```python
df.drop_duplicates()
```

```python
#order the data data a little to see
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
sns.regplot(x="gross",y="budget",data=df,scatter_kws={"color":"orange"},line_kws={"color":"black"})
```

```python
sns.regplot(x="score",y="gross",data=df,scatter_kws={"color":"orange"})
```

```python
#correlation matrix between all numeric columns
df.corr(method='pearson')
```

```python
df.corr(method="kendall")
```

```python
df.corr(method="spearman")
```

```python
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.title("correlation matrix for numerical features")
plt.xlabel("movie features")
plt.ylabel("movie features")
plt.show()
```

```python
#using factorize
df.apply(lambda x:x.factorize()[0]).corr(method='pearson')
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True)
plt.title("correlation matrix for movies")
plt.xlabel("movie features")
plt.ylabel("movie features")


plt.show()
```

```python
correlation_mat=df.apply(lambda x: x.factorize()[0]).corr()
corr_pairs=correlation_mat.unstack()
print(corr_pairs)
```

```python
sorted_pairs=corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
